### Sine equation :
$y(t) = A\cdot \sin(\omega\cdot t + \varphi)$

With :
- $A$ = Amplitude (rad)
- $\omega$ = Angular frequency (rad/sec) $(2\pi \cdot f)$
- $t$ = Time (sec)
- $\varphi$ = Phase shift (rad)
- $f$ = Frequency (number of oscillations per second)


```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0,1,1000) # time vector

# signal parameters
A = 1 # amplitude
f = 1 # frequency
phi = np.pi*2 # phase shift

# calculation
w = 2 * np.pi * f # angular frequency
sin_s = A * np.sin(w * t + phi) # signal
cos_s = A * np.cos(w * t + phi) # signal

# plot
plt.plot(t,sin_s, label= 'Sine wave')
plt.plot(t,cos_s, label= 'Cosine wave')
plt.title('Sine and Cosine Waves')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (rad)')
plt.legend()
plt.grid()
plt.show() 
```

```python
# Creation of a function for plot multiple Sine waves
t = np.linspace(0,1,1000) # time vector

# Sin function
def sinusoid(A,f,phi):
    w = 2 * np.pi * f # angular frequency
    x = A * np.sin(w * t + phi) # signal
    return x

s1 = sinusoid(1,1,(2*np.pi))
s2 = sinusoid(1,2,(2*np.pi))

# plot
plt.plot(t,s1, label= 'Signal 1')
plt.plot(t,s2, label= 'Signal 2')
plt.title('Sine and Cosine Waves')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (rad)')
plt.legend()
plt.grid()
plt.show() 
```
