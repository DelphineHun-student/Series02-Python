### Sine equation :
$y(t) = A\cdot \sin(\omega\cdot t + \varphi)$

With :
- $A$ = Amplitude (rad)
- $\omega$ = Angular frequency (rad/sec) $(2\pi \cdot f)$
- $t$ = Time (sec)
- $\varphi$ = Phase shift (rad)
- $f$ = Frequency (number of oscillations per second)


```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0,1,1000) # time vector

# signal parameters
A = 2 # amplitude
f = 1 # frequency
phi = np.pi / 4 # phase shift

# calculation
w = 2 * np.pi * f # angular frequency
s = A * np.sin(w * t + phi) # signal

# plot
plt.plot(t,s)
plt.title('Sine Wave')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (rad)')
plt.grid()

```
