### Sine equation :
$y(t) = A\cdot \sin(\omega\cdot t + \varphi)$

With :
- $A$ = Amplitude (rad)
- $\omega$ = Angular frequency (rad/sec) $(2\pi \cdot f)$
- $t$ = Time (sec)
- $\varphi$ = Phase shift (rad)
- $f$ = Frequency (number of oscillations per second)
